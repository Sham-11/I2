```python
from transformers import MBartForConditionalGeneration, MBart50TokenizerFast
```

```python
model_name = "facebook/mbart-large-50-many-to-many-mmt"
tokenizer = MBart50TokenizerFast.from_pretrained(model_name)
model = MBartForConditionalGeneration.from_pretrained(model_name)
```

```python
lang_codes = {
    "English": "en_XX",
    "Hindi": "hi_IN",
    "French": "fr_XX",
    "Spanish": "es_XX",
    "German": "de_DE",
    "Japanese": "ja_XX",
    "Chinese": "zh_CN",
    "Tamil": "ta_IN",
    "Telugu": "te_IN",
    "Marathi": "mr_IN",
    "Bengali": "bn_IN",
    "Punjabi": "pa_IN",
    "Gujarati": "gu_IN",
    "Korean": "ko_KR"
}
```

```python
print("Available languages:")
for i, lang in enumerate(lang_codes, 1):
    print(f"{i}. {lang}")
```

```python
source_lang = input("Enter source language: ").strip().title()
target_lang = input("Enter target language: ").strip().title()
text = input("Enter the text you want to translate: ").strip()

if source_lang not in lang_codes or target_lang not in lang_codes:
    print("One or both languages are not supported. Please try again.")
else:
    tokenizer.src_lang = lang_codes[source_lang]

    encoded = tokenizer(text, return_tensors="pt")
    generated = model.generate(
        **encoded,
        forced_bos_token_id=tokenizer.lang_code_to_id[lang_codes[target_lang]]
    )
    translated = tokenizer.batch_decode(generated, skip_special_tokens=True)[0]

    print(f"\nTranslated Text ({source_lang} → {target_lang}):")
    print(translated)
```
