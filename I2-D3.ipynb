```python
w1=1
w2=1
b=-1
learning_rate=0.1
```

```python
def function(x):
  if (x>=0):
    return 1
  else:
    return 0
```

```python
for epoch in range(20):
  print("Epoch ",(epoch+1))

  for x,y in zip(inputs,output):
     z = w1 * x[0] + w2 * x[1] + b
     pred = function(z)
     error = y - pred

     # Update weights and bias
     w1 += learning_rate * error * x[0]
     w2 += learning_rate * error * x[1]
     b += learning_rate * error

     print(f"Input: {x}, Predicted: {pred}, Actual: {y}, Error: {error}")
print("\n")
```

With Error:

```python
import matplotlib.pyplot as plt
import numpy as np

inputs = [[0, 0], [0, 1], [1, 0], [1, 1]]
labels = [0, 0, 0, 1]

w1, w2, b = 0.0, 0.0, 0.0
learning_rate = 0.1

def function(x):
    return 1 if x >= 0 else 0

# Train the perceptron
for epoch in range(20):
    total_error = 0
    print(f"Epoch {epoch + 1}")
    for x, y in zip(inputs, labels):
        z = w1 * x[0] + w2 * x[1] + b
        prediction = function(z)
        error = y - prediction
        total_error += abs(error)  #accumulate total error for this epoch

        # Update weights and bias
        w1 += learning_rate * error * x[0]
        w2 += learning_rate * error * x[1]
        b += learning_rate * error

        print(f"Input: {x}, Prediction: {prediction}, Actual: {y}, Error: {error}")

    print(f"Total error this epoch: {total_error}\n")

print("Final predictions after training:")
for x in inputs:
    z = w1 * x[0] + w2 * x[1] + b
    print(f"Input: {x} => Output: {function(z)}")

# Extra: Plot decision boundary
class_0 = [inputs[i] for i in range(len(inputs)) if labels[i] == 0]
class_1 = [inputs[i] for i in range(len(inputs)) if labels[i] == 1]

x0 = [i[0] for i in class_0]
y0 = [i[1] for i in class_0]
x1 = [i[0] for i in class_1]
y1 = [i[1] for i in class_1]

plt.scatter(x0, y0, color='red', label='Class 0')
plt.scatter(x1, y1, color='green', label='Class 1')

x_vals = np.linspace(-0.5, 1.5, 100)
if w2 != 0:
    y_vals = -(w1 * x_vals + b) / w2
    plt.plot(x_vals, y_vals, label='Decision Boundary')
else:
    plt.axvline(x=-b/w1, label='Decision Boundary')

plt.xlabel('x1')
plt.ylabel('x2')
plt.title('Perceptron Decision Boundary for AND Gate')
plt.legend()
plt.grid(True)
plt.show()
```

Multi-Layer Perceptron XOR

```python
import numpy as np

# Sigmoid activation and its derivative
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def sigmoid_derivative(x):
    return x * (1 - x)
```

```python
inputs = np.array([[0, 0],[0, 1],[1, 0],[1, 1]])
labels = np.array([0,1,1,0])
```

```python
# Seed for reproducibility
np.random.seed(42)

# Initialize weights and biases
input_layer_neurons = 2
hidden_layer_neurons = 2
output_neurons = 1
```

```python
# Weights
wh = np.random.uniform(size=(input_layer_neurons, hidden_layer_neurons))  # weights input -> hidden
bh = np.random.uniform(size=(1, hidden_layer_neurons))                    # bias for hidden
wo = np.random.uniform(size=(hidden_layer_neurons, output_neurons))       # weights hidden -> output
bo = np.random.uniform(size=(1, output_neurons))
```

```python
# Training
learning_rate = 0.1
epochs = 10000

for epoch in range(epochs):
    # Forward pass
    hidden_input = np.dot(inputs, wh) + bh
    hidden_output = sigmoid(hidden_input)

    final_input = np.dot(hidden_output, wo) + bo
    predicted_output = sigmoid(final_input)

    # Backpropagation
    error = labels - predicted_output
    d_output = error * sigmoid_derivative(predicted_output)

    labels = labels.reshape(-1, 1)
    predicted_output = predicted_output.reshape(-1, 1)

    error_hidden = d_output.dot(wo.T)
    d_hidden = error_hidden * sigmoid_derivative(hidden_output)

    # Update weights and biases
    wo += hidden_output.T.dot(d_output) * learning_rate
    bo += np.sum(d_output, axis=0, keepdims=True) * learning_rate
    wh += inputs.T.dot(d_hidden) * learning_rate
    bh += np.sum(d_hidden, axis=0, keepdims=True) * learning_rate

    # Optional: print error every 1000 epochs
    if epoch % 1000 == 0:
        loss = np.mean(np.abs(error))
        print(f"Epoch {epoch} - Loss: {loss:.4f}")

# Final output
print("\nPredicted outputs after training:")
print(np.round(predicted_output, 3))

print("\nExpected XOR outputs:")
print(labels)
```

```python
import matplotlib.pyplot as plt

# Generate a grid of points over the input space
xx, yy = np.meshgrid(np.linspace(0, 1, 100),
                     np.linspace(0, 1, 100))
grid = np.c_[xx.ravel(), yy.ravel()]

# Forward pass through trained MLP
hidden = sigmoid(np.dot(grid, wh) + bh)
output = sigmoid(np.dot(hidden, wo) + bo)
output = output.reshape(xx.shape)

# Plot decision boundary
plt.contourf(xx, yy, output, levels=[0, 0.5, 1], colors=['#ffcccc', '#ccffcc'])
plt.colorbar()
plt.title("MLP Decision Boundary for XOR")
plt.xlabel("Input 1")
plt.ylabel("Input 2")

# Plot original data points
for i in range(len(inputs)):
    x, y = inputs[i]
    label = labels[i][0]
    plt.scatter(x, y, c='black' if label == 0 else 'blue', edgecolors='white', s=100)

plt.show()
```
